El F1-score es una métrica adecuada para evaluar el desempeño de modelos de clasificación, especialmente cuando existe desbalance entre clases.

* El promediado macro calcula el promedio del F1-score de cada clase, otorgando el mismo peso a todas las clases.
* El promediado micro calcula las métricas de forma global considerando todas las predicciones; en problemas de clasificación multiclase suele ser equivalente a la accuracy, por lo que no es la mejor métrica cuando el dataset está desbalanceado.

```python
f1_score(y_test, y_pred, average='macro')
```

## **Consigna del Desafío 1**
**Cada experimento realizado debe estar acompañado de una explicación o interpretación de lo observado.**



**1. Vectorizar documentos**

Estudiar los 5 documentos más similares de cada uno analizar si tiene sentido
la similaridad según el contenido del texto y la etiqueta de clasificación.

**2. Construir un modelo de clasificación por prototipos (tipo zero-shot).**
* Clasificar los documentos de un conjunto de test comparando cada uno con todos los de entrenamiento y asignar la clase al label del documento del conjunto de entrenamiento con mayor similaridad.

**3. Entrenar modelos de clasificación Naïve Bayes para maximizar el desempeño de clasificación**

* F1-Score Macro en el conjunto de datos de test. Considerar cambiar parámetros
de instanciación del vectorizador y los modelos y probar modelos de Naïve Bayes Multinomial y ComplementNB.

**NO cambiar el hiperparámetro ngram_range de los vectorizadores**.

**4. Transponer la matriz documento-término.**
* De esa manera se obtiene una matriz término-documento que puede ser interpretada como una colección de vectorización de palabras.
* Estudiar ahora similaridad entre palabras tomando 5 palabras y estudiando sus 5 más similares.

**Elegir las palabras MANUALMENTE para evitar la aparición de términos poco interpretables**.


```python
# ================================
# DESAFÍO 1 - RESOLUCIÓN COMPLETA
# Procesamiento de Lenguaje Natural I
# ================================

import numpy as np
import pandas as pd

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.metrics import f1_score, accuracy_score, classification_report
from sklearn.model_selection import ParameterGrid

# Para reproducibilidad
RANDOM_STATE = 42
rng = np.random.default_rng(RANDOM_STATE)

# --------------------------------------------------
# 0) CARGA DE DATOS
# --------------------------------------------------
newsgroups_train = fetch_20newsgroups(
    subset='train',
    remove=('headers', 'footers', 'quotes')
)

newsgroups_test = fetch_20newsgroups(
    subset='test',
    remove=('headers', 'footers', 'quotes')
)

y_train = newsgroups_train.target
y_test = newsgroups_test.target
target_names = newsgroups_train.target_names

print(f"Cantidad de documentos de entrenamiento: {len(newsgroups_train.data)}")
print(f"Cantidad de documentos de test: {len(newsgroups_test.data)}")
print(f"Cantidad de clases: {len(target_names)}")
print(target_names)

```

## 1. Vectorización de documentos y similaridad entre documentos

Para representar los documentos se utiliza **TF-IDF**, que pondera cada término según su frecuencia en el documento y su rareza dentro del corpus. Luego, la similaridad entre documentos se mide con **cosine similarity**.

La idea es que dos documentos serán similares si comparten vocabulario relevante y si sus vectores TF-IDF apuntan en una dirección parecida dentro del espacio vectorial.


```python
# --------------------------------------------------
# 1) VECTORIZACIÓN TF-IDF
# --------------------------------------------------

# No se modifica ngram_range, por lo tanto queda en su valor default: (1, 1)
tfidfvect = TfidfVectorizer()

X_train = tfidfvect.fit_transform(newsgroups_train.data)
X_test = tfidfvect.transform(newsgroups_test.data)

idx2word = {v: k for k, v in tfidfvect.vocabulary_.items()}

print(f"Shape de X_train: {X_train.shape}")
print(f"Shape de X_test: {X_test.shape}")
print(f"Tamaño del vocabulario: {len(tfidfvect.vocabulary_)}")

```

```python
# --------------------------------------------------
# 1) TOMAR 5 DOCUMENTOS AL AZAR Y BUSCAR SUS 5 MÁS SIMILARES
# --------------------------------------------------

# Elegimos 5 documentos aleatorios de train
random_docs = rng.choice(X_train.shape[0], size=5, replace=False)

similarity_results = []

for doc_idx in random_docs:
    cossim = cosine_similarity(X_train[doc_idx], X_train)[0]
    most_similar = np.argsort(cossim)[::-1][1:6]  # excluimos el mismo documento
    
    print("=" * 100)
    print(f"DOCUMENTO BASE - índice: {doc_idx}")
    print(f"Clase real: {target_names[y_train[doc_idx]]}")
    print("Texto base, primeros 700 caracteres:")
    print(newsgroups_train.data[doc_idx][:700].replace("\n", " "))
    print("\nTop 5 documentos más similares:")
    
    for rank, sim_idx in enumerate(most_similar, start=1):
        base_class = target_names[y_train[doc_idx]]
        sim_class = target_names[y_train[sim_idx]]
        same_class = base_class == sim_class
        
        similarity_results.append({
            "documento_base": doc_idx,
            "clase_base": base_class,
            "rank": rank,
            "documento_similar": sim_idx,
            "clase_similar": sim_class,
            "cosine_similarity": cossim[sim_idx],
            "misma_clase": same_class
        })
        
        print(f"\n{rank}) índice: {sim_idx}")
        print(f"   Similaridad coseno: {cossim[sim_idx]:.4f}")
        print(f"   Clase: {sim_class}")
        print(f"   ¿Misma clase?: {same_class}")
        print(f"   Texto, primeros 350 caracteres: {newsgroups_train.data[sim_idx][:350].replace(chr(10), ' ')}")

similarity_df = pd.DataFrame(similarity_results)
similarity_df

```

### Interpretación del punto 1

En general, los documentos más similares tienden a pertenecer a la misma clase o a clases temáticamente cercanas. Esto ocurre porque TF-IDF asigna mayor peso a términos representativos del contenido. Por ejemplo, documentos sobre computadoras suelen compartir términos técnicos; documentos sobre religión comparten vocabulario doctrinal; y documentos sobre autos o motos comparten términos mecánicos.

Sin embargo, no siempre los documentos más similares pertenecen exactamente a la misma etiqueta. Esto es esperable porque la similaridad TF-IDF mide coincidencia léxica, no comprensión semántica profunda. Dos publicaciones de clases distintas pueden compartir términos, estilo o contexto, especialmente en categorías cercanas.


## 2. Modelo de clasificación por prototipos usando similaridad coseno

Se construye un clasificador tipo **nearest prototype / nearest neighbor semántico**. Para cada documento de test se calcula la similaridad contra todos los documentos de entrenamiento. Luego se asigna como predicción la clase del documento de entrenamiento más similar.

No hay entrenamiento paramétrico del modelo: la clasificación depende directamente de la representación TF-IDF y de la similaridad coseno.


```python
# --------------------------------------------------
# 2) CLASIFICADOR POR PROTOTIPOS / DOCUMENTO MÁS SIMILAR
# --------------------------------------------------

# cosine_similarity(X_test, X_train) produce una matriz de tamaño:
# n_documentos_test x n_documentos_train
similarities_test_train = cosine_similarity(X_test, X_train)

# Para cada documento de test, buscamos el documento de train más similar
nearest_train_idx = np.argmax(similarities_test_train, axis=1)

# La clase predicha es la clase del documento de train más parecido
y_pred_proto = y_train[nearest_train_idx]

proto_accuracy = accuracy_score(y_test, y_pred_proto)
proto_f1_macro = f1_score(y_test, y_pred_proto, average='macro')

print(f"Accuracy del clasificador por prototipos: {proto_accuracy:.4f}")
print(f"F1 macro del clasificador por prototipos: {proto_f1_macro:.4f}")

print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred_proto, target_names=target_names))

```

### Interpretación del punto 2

El clasificador por prototipos funciona como una clasificación basada en memoria: no aprende parámetros, sino que asigna la clase del documento de entrenamiento más parecido. Su desempeño depende fuertemente de que exista en train algún documento muy similar al de test.

Este enfoque suele ser razonable como línea base, pero puede fallar cuando el documento de test usa vocabulario distinto, cuando hay ruido o cuando dos clases comparten terminología. A diferencia de Naïve Bayes, no aprende patrones globales por clase; se apoya en comparaciones documento a documento.


## 3. Modelos Naïve Bayes y optimización de desempeño

Se comparan dos variantes:

- **MultinomialNB**: modelo clásico para conteos o representaciones TF-IDF en clasificación de texto.
- **ComplementNB**: variante especialmente útil en problemas de texto y datasets desbalanceados, porque estima los pesos usando información complementaria de cada clase.

Se prueban distintos parámetros del vectorizador y del modelo. Se respeta la restricción de la consigna: **no se modifica `ngram_range`**, por lo que se mantiene en su valor por defecto `(1, 1)`.


```python
# --------------------------------------------------
# 3) OPTIMIZACIÓN DE NAÏVE BAYES
# --------------------------------------------------

experiments = []

vectorizer_grid = list(ParameterGrid({
    # No se incluye ngram_range porque la consigna pide NO modificarlo
    "max_df": [0.7, 0.85, 1.0],
    "min_df": [1, 2, 5],
    "max_features": [None, 30000, 50000],
    "stop_words": [None, "english"],
    "sublinear_tf": [False, True],
    "use_idf": [True],
    "norm": ["l2"]
}))

model_configs = []

for alpha in [0.01, 0.05, 0.1, 0.5, 1.0]:
    model_configs.append(("MultinomialNB", MultinomialNB(alpha=alpha), {"alpha": alpha}))
    model_configs.append(("ComplementNB", ComplementNB(alpha=alpha), {"alpha": alpha}))

for vect_params in vectorizer_grid:
    vect = TfidfVectorizer(**vect_params)
    
    Xtr = vect.fit_transform(newsgroups_train.data)
    Xte = vect.transform(newsgroups_test.data)
    
    for model_name, model, model_params in model_configs:
        model.fit(Xtr, y_train)
        pred = model.predict(Xte)
        
        f1_macro = f1_score(y_test, pred, average="macro")
        acc = accuracy_score(y_test, pred)
        
        experiments.append({
            "model": model_name,
            "f1_macro": f1_macro,
            "accuracy": acc,
            **model_params,
            **vect_params,
            "vocab_size": Xtr.shape[1]
        })

results_df = pd.DataFrame(experiments).sort_values("f1_macro", ascending=False).reset_index(drop=True)
results_df.head(15)

```

```python
# Mejor configuración encontrada
best_result = results_df.iloc[0]
best_result

```

```python
# Reentrenamiento del mejor modelo para obtener reporte detallado
best_vect_params = {
    "max_df": best_result["max_df"],
    "min_df": int(best_result["min_df"]),
    "max_features": None if pd.isna(best_result["max_features"]) else int(best_result["max_features"]),
    "stop_words": None if pd.isna(best_result["stop_words"]) else best_result["stop_words"],
    "sublinear_tf": bool(best_result["sublinear_tf"]),
    "use_idf": bool(best_result["use_idf"]),
    "norm": best_result["norm"]
}

best_vectorizer = TfidfVectorizer(**best_vect_params)
Xtr_best = best_vectorizer.fit_transform(newsgroups_train.data)
Xte_best = best_vectorizer.transform(newsgroups_test.data)

if best_result["model"] == "MultinomialNB":
    best_model = MultinomialNB(alpha=float(best_result["alpha"]))
else:
    best_model = ComplementNB(alpha=float(best_result["alpha"]))

best_model.fit(Xtr_best, y_train)
y_pred_best = best_model.predict(Xte_best)

print("Mejor modelo encontrado:")
print(best_result[["model", "alpha", "f1_macro", "accuracy", "max_df", "min_df", "max_features", "stop_words", "sublinear_tf", "vocab_size"]])

print("\nReporte de clasificación del mejor modelo:")
print(classification_report(y_test, y_pred_best, target_names=target_names))

```

```python
# Tabla comparativa resumida entre prototipos y mejores Naïve Bayes
best_by_model = results_df.groupby("model", as_index=False).first()[["model", "f1_macro", "accuracy", "alpha", "max_df", "min_df", "max_features", "stop_words", "sublinear_tf", "vocab_size"]]

comparison_df = pd.concat([
    pd.DataFrame([{
        "model": "Clasificador por prototipos",
        "f1_macro": proto_f1_macro,
        "accuracy": proto_accuracy,
        "alpha": None,
        "max_df": None,
        "min_df": None,
        "max_features": None,
        "stop_words": None,
        "sublinear_tf": None,
        "vocab_size": X_train.shape[1]
    }]),
    best_by_model
], ignore_index=True).sort_values("f1_macro", ascending=False)

comparison_df

```

### Interpretación del punto 3

Los modelos Naïve Bayes suelen superar al clasificador por prototipos porque aprenden una distribución global de términos por clase. En vez de depender de un único documento parecido, estiman qué términos son más característicos de cada categoría.

`ComplementNB` suele obtener muy buen rendimiento en clasificación de texto porque corrige algunas limitaciones de `MultinomialNB`, especialmente cuando hay clases con vocabulario parcialmente solapado o distribuciones no perfectamente balanceadas.

Los parámetros del vectorizador también influyen considerablemente. Por ejemplo, `min_df` elimina términos extremadamente raros, `max_df` reduce términos demasiado frecuentes y poco discriminativos, `stop_words='english'` puede mejorar el rendimiento al quitar palabras funcionales del inglés. Sin embargo, si se eliminan demasiados términos, también puede perderse información útil.


## 4. Similaridad entre palabras transponiendo la matriz documento-término

La matriz original `X_train` tiene forma:

- filas = documentos
- columnas = términos

Al transponerla, se obtiene una matriz término-documento:

- filas = términos
- columnas = documentos

De esta manera, cada palabra queda representada por los documentos en los que aparece y con qué peso TF-IDF. Dos palabras serán similares si tienden a aparecer en documentos parecidos.


```python
# --------------------------------------------------
# 4) TRANSPONER MATRIZ DOCUMENTO-TÉRMINO Y ANALIZAR SIMILARIDAD ENTRE PALABRAS
# --------------------------------------------------

# Usamos el vectorizador base del punto 1
X_terms = X_train.T
feature_names = np.array(tfidfvect.get_feature_names_out())

manual_words = ["space", "car", "god", "windows", "baseball"]

word_similarity_results = []

for word in manual_words:
    if word not in tfidfvect.vocabulary_:
        print(f"La palabra '{word}' no está en el vocabulario.")
        continue
    
    word_idx = tfidfvect.vocabulary_[word]
    cossim_words = cosine_similarity(X_terms[word_idx], X_terms)[0]
    most_similar_word_idx = np.argsort(cossim_words)[::-1][1:6]
    
    print("=" * 80)
    print(f"Palabra base: {word}")
    print("Top 5 palabras más similares:")
    
    for rank, sim_word_idx in enumerate(most_similar_word_idx, start=1):
        similar_word = feature_names[sim_word_idx]
        score = cossim_words[sim_word_idx]
        
        word_similarity_results.append({
            "palabra_base": word,
            "rank": rank,
            "palabra_similar": similar_word,
            "cosine_similarity": score
        })
        
        print(f"{rank}) {similar_word:20s} similarity={score:.4f}")

word_similarity_df = pd.DataFrame(word_similarity_results)
word_similarity_df

```

### Interpretación del punto 4

La similaridad entre palabras no se calcula por significado lingüístico directo, sino por **distribución de aparición en documentos**. Es decir, dos palabras aparecen como similares si suelen estar presentes en documentos parecidos.

Por eso, palabras como `space` pueden aparecer cerca de términos relacionados con ciencia, NASA u órbitas; `car` puede asociarse con términos mecánicos; `god` con discusiones religiosas; `windows` con informática; y `baseball` con deportes.

Este enfoque es una forma simple de semántica distribucional: las palabras se caracterizan por los contextos documentales en los que aparecen. Aun así, tiene limitaciones porque depende del corpus, del vocabulario y de la frecuencia de aparición.


## Conclusión general

En este desafío se observa que TF-IDF permite transformar texto no estructurado en vectores numéricos comparables. La similaridad coseno resulta útil para encontrar documentos o palabras relacionadas, aunque se basa principalmente en coincidencias léxicas y patrones de coocurrencia.

El clasificador por prototipos funciona como una línea base basada en similaridad, pero los modelos Naïve Bayes suelen ser más robustos porque aprenden patrones globales por clase. Entre las variantes evaluadas, `ComplementNB` frecuentemente resulta competitivo o superior en problemas de clasificación de texto.

Finalmente, al transponer la matriz documento-término es posible estudiar relaciones entre palabras según los documentos en los que aparecen, lo cual introduce la idea de representación distribucional del significado.
